```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
penguins = pd.read_csv("penguins.csv")
sns.barplot(x="species",y="body_mass_g",data=penguins, hue="sex")
#sns.pairplot(penguins, hue="species")
#sns.regplot(x="flipper_length_mm",y="body_mass_g",data=penguins)
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
df = pd.read_excel("data1.xlsx")
plt.plot(df["kor"], "k--", marker="o", color="g", label='Kor score')
plt.plot(df["math"],color="r",marker="^",label="Math score")
plt.title('score graph')
plt.xlabel('name')
plt.ylabel('score')
plt.legend()
plt.show()
```
